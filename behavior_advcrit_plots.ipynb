```python
def perc_over_crit(total, list_of_labels):
    f, (ax1) = plt.subplots(1, 1, figsize = (5,4))
    ax1.plot(total)
    ax1.set_title('% of trials over advancement criterion')
    ax1.set_ylabel('% of tirals over advancement criterion')
    #ax1.set_xlabel('session')
    plt.locator_params(axis='x', nbins=5)
    plt.locator_params(axis='y', nbins=5)
    ticks = list(range(len(total)))
    ax1.set_xticks(ticks, list_of_labels, fontsize = 10)
    plt.yticks(fontsize = 10)
    f.tight_layout()
    
def peak_avg(total, list_of_labels):
    f, (ax1) = plt.subplots(1, 1, figsize = (5,4))
    ax1.plot(total)
    ax1.set_title('Peak moving average of performance over 30 trials')
    ax1.set_ylabel('moving average')
    #ax1.set_xlabel('session')
    plt.axhline(y = 0.8, color = 'b', linestyle = '--')
    ticks = list(range(len(total)))
    ax1.set_xticks(ticks, list_of_labels, fontsize = 10)
    plt.yticks(fontsize = 10)
    plt.locator_params(axis='x', nbins=8)
    plt.locator_params(axis='y', nbins=4)
    f.tight_layout()
```

```python
total = [.2473, .1762, .4652, 0]
list_of_labels = ['7/12', '7/13', '7/14', '7/17']
perc_over_crit(total, list_of_labels)
```

```python
total = [.933, .9, .9, .78]
list_of_labels = ['7/12', '7/13', '7/14', '7/17']
peak_avg(total, list_of_labels)
```
